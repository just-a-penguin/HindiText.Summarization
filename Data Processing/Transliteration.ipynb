```python
import pandas as pd
import re
from ai4bharat.transliteration import XlitEngine
# Load the CSV file
df = pd.read_csv('hindi_train.csv')


# Initialize the transliteration engine for Hindi with a specified beam width
translit_engine = XlitEngine("hi", beam_width=15)
transliteration_cache = {'www': 'डब्ल्यूडब्ल्यूडब्ल्यू'}
# Function to transliterate only English words in the text
def transliterate_english_parts(text):
    # Regular expression to find words or sequences that include English alphabets (may include numbers)
    parts = re.findall(r'\b[a-zA-Z0-9]*[a-zA-Z][a-zA-Z0-9]*\b', text)
    for part in parts:
        # Check cache first to avoid repeated transliteration
        if part not in transliteration_cache:
            try:
                # Transliterate and store in cache
                transliterated_part = translit_engine.translit_sentence(part)['hi']
                transliteration_cache[part] = transliterated_part
            except Exception as e:
                print(f"Error transliterating part {part}: {e}")
                transliteration_cache[part] = part  # Use the original part if an error occurs
        # Replace the original part with its Hindi transliteration in the text
        text = text.replace(part, transliteration_cache[part])
    return text
# Apply the function to 'Article' and 'Heading' columns
df['Article'] = df['Article'].apply(transliterate_english_parts)
df['Heading'] = df['Heading'].apply(transliterate_english_parts)
df['Summary'] = df['Summary'].apply(transliterate_english_parts)
# Save the modified dataframe to a new CSV file
df.to_csv('Translated_hindi_train.csv', index=False)
```
