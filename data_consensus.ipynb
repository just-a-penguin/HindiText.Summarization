```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('2023_data/hindi_train.csv')  
```

```python
articles = df_train['Article'].values
summaries = df_train['Summary'].values
```

```python
print(articles[0])
```

```python
#give code to graph the distribution of the lengths of the articles and summaries
#you can use matplotlib or seaborn

#graph the distribution of the lengths of the articles
import matplotlib.pyplot as plt
import seaborn as sns

article_lengths = [len(article.split()) for article in articles]
summary_lengths = [len(summary.split()) for summary in summaries]

sns.displot(article_lengths, kde=False)
plt.title('Article Lengths')
plt.xlabel('Article Length')
plt.ylabel('Frequency')
plt.show()
```
